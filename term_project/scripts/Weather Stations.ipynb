```python
## GMT TO EST  = pytz: World Timezone Definitions for Python 

import pytz                # http://pytz.sourceforge.net/
from pytz import timezone
from datetime import datetime

gmt = pytz.timezone('GMT')
eastern = pytz.timezone('US/Eastern')
time = "201501010000"
dategmt = gmt.localize(datetime.strptime(time, '%Y%m%d%H%M'))
dateeastern = dategmt.astimezone(eastern)
print dategmt
print dateeastern
```

```python
def change_time(time):
    gmt = pytz.timezone('GMT')
    eastern = pytz.timezone('US/Eastern')
    dategmt = gmt.localize(datetime.strptime(str(time), '%Y%m%d%H%M'))
    dateeastern = dategmt.astimezone(eastern)
    return dateeastern

weather['YR--MODAHRMN'] = weather['YR--MODAHRMN'].apply(change_time)
weather.head(10)
```

```python
import datetime
weather['datetime'] =map(lambda x: x.strftime('%Y-%m-%d %H:%M:%S') ,weather['YR--MODAHRMN'] )
#weather['time']=weather['YR--MODAHRMN']
```

```python
weather_tofile = weather[ ['datetime','SPD','TEMP','SD','PCP01'] ]
weather_tofile.tail(5)
```
